```python
candlestick = go.Candlestick(x = price_data['time'], close=price_data['close'], open=price_data['open'], high=price_data['high'], low=price_data['low'])
```

```python
# Create a candlestick figure   
fig = go.Figure(data=[candlestick])
fig.update_layout(title='XAUUSD prices')                        

# Show the plot
fig.show()
```
